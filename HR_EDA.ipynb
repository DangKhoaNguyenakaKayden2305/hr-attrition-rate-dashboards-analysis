```python
# ============================================================
# Step 1: Import required libraries
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Display settings for easier inspection
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)
pd.set_option("display.float_format", "{:,.2f}".format)
```

```python
# ============================================================
# Step 2: Load the Excel dataset
# ============================================================

# Update this path depending on where you store the file
file_path = r"C:\Users\Kaydxn Nguyen\Documents\ASM2_HR_project\Assessment 2 Case Study Pacakge (updated)\updated_hr_dataset_v2.xlsx"

# If running locally and file is in the same folder, use:
# file_path = "updated_hr_dataset_v2.xlsx"

# Load workbook
excel_file = pd.ExcelFile(file_path)

# Check available sheets
print("Available sheets:")
print(excel_file.sheet_names)

# Load main HR dataset
df = pd.read_excel(file_path, sheet_name="Updated HR Dataset")

# Preview dataset shape
print("\nDataset shape:")
print(df.shape)

# Preview first 5 rows
df.head()
```

```python
# ============================================================
# Step 3: Inspect columns and data structure
# ============================================================

print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])

print("\nColumn names:")
for col in df.columns:
    print("-", col)

print("\nData types:")
print(df.dtypes)

print("\nBasic dataset preview:")
display(df.head())
```

```python
# ============================================================
# Step 4: Check missing values
# ============================================================

missing_summary = (
    df.isna()
    .sum()
    .reset_index()
    .rename(columns={"index": "Column", 0: "Missing Count"})
)

missing_summary["Missing %"] = (
    missing_summary["Missing Count"] / len(df) * 100
).round(2)

missing_summary = missing_summary.sort_values(
    by="Missing Count", ascending=False
)

display(missing_summary)
```

```python
# ============================================================
# Step 5: Check duplicate employee records
# ============================================================

duplicate_employee_id = df["Employee ID"].duplicated().sum()
duplicate_employee_number = df["Employee Number"].duplicated().sum()

print("Duplicate Employee ID:", duplicate_employee_id)
print("Duplicate Employee Number:", duplicate_employee_number)

# Show duplicated rows if any
duplicates = df[df["Employee ID"].duplicated(keep=False)]

if len(duplicates) > 0:
    display(duplicates.sort_values("Employee ID"))
else:
    print("No duplicated Employee ID found.")
```

```python
# ============================================================
# Step 6: Standardise column names for Python analysis
# ============================================================

df_clean = df.copy()

# Function to convert column names to snake_case
def clean_column_name(col):
    col = col.strip()
    col = col.lower()
    col = col.replace("(", "")
    col = col.replace(")", "")
    col = col.replace("/", "_")
    col = col.replace("-", "_")
    col = col.replace("&", "and")
    col = col.replace(" ", "_")
    col = col.replace("__", "_")
    return col

df_clean.columns = [clean_column_name(col) for col in df_clean.columns]

print("Cleaned column names:")
for col in df_clean.columns:
    print("-", col)
```

```python
# ============================================================
# Step 7: Rename long columns into practical names
# ============================================================

rename_dict = {
    "identifies_as_aboriginal_and_torres_strait_islander": "atsi_identification",
    "yearly_base_salary_aud": "base_salary_aud",
    "oncosts_aud": "oncosts_aud",
    "super_aud": "super_aud",
    "flexible_work_arrangement": "flexible_work",
    "current_employee_status": "employee_status"
}

df_clean = df_clean.rename(columns=rename_dict)

print("Updated columns:")
print(df_clean.columns.tolist())
```

```python
# ============================================================
# Step 8: Validate key binary fields
# ============================================================

binary_fields = [
    "hired_flag",
    "terminated_flag",
    "active_employee"
]

for field in binary_fields:
    print(f"\nValue counts for {field}:")
    print(df_clean[field].value_counts(dropna=False))
```

```python
# ============================================================
# Step 9: Validate ordinal 1-5 people metrics
# ============================================================

scale_fields = [
    "inclusion_index",
    "wellbeing_index",
    "development_index",
    "work_life_balance",
    "job_satisfaction",
    "performance_rating",
    "sdg_score"
]

for field in scale_fields:
    print(f"\n{field}")
    print("Min:", df_clean[field].min())
    print("Max:", df_clean[field].max())
    print("Value counts:")
    print(df_clean[field].value_counts().sort_index())
```

```python
# ============================================================
# Step 10: Create total employment cost
# ============================================================

df_clean["total_employment_cost_aud"] = (
    df_clean["base_salary_aud"]
    + df_clean["oncosts_aud"]
    + df_clean["super_aud"]
)

# Check result
df_clean[[
    "base_salary_aud",
    "oncosts_aud",
    "super_aud",
    "total_employment_cost_aud"
]].head()
```

```python
# ============================================================
# Step 11: Create attrition support fields
# ============================================================

df_clean["is_terminated"] = np.where(df_clean["terminated_flag"] == 1, 1, 0)
df_clean["is_active"] = np.where(df_clean["active_employee"] == 1, 1, 0)

overall_attrition_rate = df_clean["is_terminated"].sum() / df_clean["hired_flag"].sum()

print("Total employees:", df_clean["hired_flag"].sum())
print("Active employees:", df_clean["is_active"].sum())
print("Terminated employees:", df_clean["is_terminated"].sum())
print("Overall attrition rate:", round(overall_attrition_rate * 100, 2), "%")
```

```python
# ============================================================
# Step 12: Create SDG band
# ============================================================

def classify_sdg(score):
    if score >= 4:
        return "High SDG"
    elif score == 3:
        return "Moderate SDG"
    else:
        return "Low SDG"

df_clean["sdg_band"] = df_clean["sdg_score"].apply(classify_sdg)

print(df_clean["sdg_band"].value_counts())
```

```python
# ============================================================
# Step 13: Create performance band
# ============================================================

def classify_performance(score):
    if score >= 4:
        return "High Performer"
    elif score == 3:
        return "Moderate Performer"
    else:
        return "Low Performer"

df_clean["performance_band"] = df_clean["performance_rating"].apply(classify_performance)

print(df_clean["performance_band"].value_counts())
```

```python
# ============================================================
# Step 14: Create satisfaction band
# ============================================================

def classify_satisfaction(score):
    if score >= 4:
        return "High Satisfaction"
    elif score == 3:
        return "Moderate Satisfaction"
    else:
        return "Low Satisfaction"

df_clean["satisfaction_band"] = df_clean["job_satisfaction"].apply(classify_satisfaction)

print(df_clean["satisfaction_band"].value_counts())
```

```python
# ============================================================
# Step 15: Create wellbeing band
# ============================================================

def classify_wellbeing(score):
    if score >= 4:
        return "High Wellbeing"
    elif score == 3:
        return "Moderate Wellbeing"
    else:
        return "Low Wellbeing"

df_clean["wellbeing_band"] = df_clean["wellbeing_index"].apply(classify_wellbeing)

print(df_clean["wellbeing_band"].value_counts())
```

```python
# ============================================================
# Step 16: Create promotion delay band
# ============================================================

def classify_promotion_delay(years):
    if years <= 1:
        return "Recently Promoted"
    elif years <= 3:
        return "Moderate Promotion Delay"
    else:
        return "Long Promotion Delay"

df_clean["promotion_delay_band"] = df_clean["years_since_last_promotion"].apply(classify_promotion_delay)

print(df_clean["promotion_delay_band"].value_counts())
```

```python
# ============================================================
# Step 17: Create tenure band
# ============================================================

def classify_tenure(years):
    if years <= 1:
        return "0-1 Years"
    elif years <= 3:
        return "2-3 Years"
    elif years <= 5:
        return "4-5 Years"
    elif years <= 10:
        return "6-10 Years"
    else:
        return "10+ Years"

df_clean["tenure_band"] = df_clean["years_at_company"].apply(classify_tenure)

print(df_clean["tenure_band"].value_counts())
```

```python
# ============================================================
# Step 18: Create comparable role group for pay equity analysis
# ============================================================

df_clean["comparable_role_group"] = (
    df_clean["department"].astype(str)
    + " | "
    + df_clean["job_role"].astype(str)
    + " | Level "
    + df_clean["job_level"].astype(str)
)

df_clean[[
    "department",
    "job_role",
    "job_level",
    "comparable_role_group"
]].head()
```

```python
# ============================================================
# Step 19: Create simple retention risk segment
# ============================================================

df_clean["retention_risk_segment"] = np.where(
    (df_clean["job_satisfaction"] <= 2) &
    (df_clean["years_since_last_promotion"] >= 4),
    "Potential Retention Risk",
    "Lower Observable Risk"
)

print(df_clean["retention_risk_segment"].value_counts())
```

```python
# ============================================================
# Step 20: Create high-value at-risk segment
# ============================================================

df_clean["high_value_at_risk"] = np.where(
    (df_clean["performance_rating"] >= 4) &
    (df_clean["job_satisfaction"] <= 2),
    "High Performer with Low Satisfaction",
    "Other"
)

print(df_clean["high_value_at_risk"].value_counts())
```

```python
# ============================================================
# Step 21: Create HQ vs Branch indicator
# ============================================================

df_clean["hq_branch_group"] = np.where(
    df_clean["office_type"].str.lower() == "hq",
    "HQ",
    "Branch"
)

print(df_clean["hq_branch_group"].value_counts())
```

```python
# ============================================================
# Step 22: Validate SDG Score formula
# ============================================================

df_clean["sdg_score_recalculated"] = np.round(
    0.35 * df_clean["inclusion_index"]
    + 0.25 * df_clean["wellbeing_index"]
    + 0.20 * df_clean["development_index"]
    + 0.20 * df_clean["job_satisfaction"]
).astype(int)

df_clean["sdg_score_match"] = np.where(
    df_clean["sdg_score"] == df_clean["sdg_score_recalculated"],
    "Match",
    "Mismatch"
)

print(df_clean["sdg_score_match"].value_counts())

# Show mismatches if any
sdg_mismatch = df_clean[df_clean["sdg_score_match"] == "Mismatch"]

if len(sdg_mismatch) > 0:
    display(sdg_mismatch[[
        "employee_id",
        "inclusion_index",
        "wellbeing_index",
        "development_index",
        "job_satisfaction",
        "sdg_score",
        "sdg_score_recalculated"
    ]].head())
else:
    print("All SDG Scores match the documented formula.")
```

```python
# ============================================================
# Step 23: Create executive summary table
# ============================================================

executive_summary = pd.DataFrame({
    "Metric": [
        "Total Employee Records",
        "Active Employees",
        "Terminated Employees",
        "Attrition Rate (%)",
        "Average Base Salary (AUD)",
        "Total Employment Cost (AUD)",
        "Average Job Satisfaction",
        "Average Wellbeing Index",
        "Average Inclusion Index",
        "Average Development Index",
        "Average SDG Score"
    ],
    "Value": [
        len(df_clean),
        df_clean["is_active"].sum(),
        df_clean["is_terminated"].sum(),
        round(overall_attrition_rate * 100, 2),
        round(df_clean["base_salary_aud"].mean(), 2),
        round(df_clean["total_employment_cost_aud"].sum(), 2),
        round(df_clean["job_satisfaction"].mean(), 2),
        round(df_clean["wellbeing_index"].mean(), 2),
        round(df_clean["inclusion_index"].mean(), 2),
        round(df_clean["development_index"].mean(), 2),
        round(df_clean["sdg_score"].mean(), 2)
    ]
})

display(executive_summary)
```

```python
# ============================================================
# Step 24: Export Tableau-ready dataset
# ============================================================

# Create processed data folder if it does not exist
processed_folder = "data/processed"
os.makedirs(processed_folder, exist_ok=True)

output_path = os.path.join(processed_folder, "hr_tableau_ready.csv")

df_clean.to_csv(output_path, index=False)

print(f"Tableau-ready dataset exported to: {output_path}")
print("Exported shape:", df_clean.shape)
```

```python
# ============================================================
# Step 25: Save summary tables for documentation
# ============================================================

summary_folder = "outputs/summary_tables"
os.makedirs(summary_folder, exist_ok=True)

# Workforce by department
department_summary = (
    df_clean
    .groupby("department")
    .agg(
        employee_count=("employee_id", "count"),
        active_employees=("is_active", "sum"),
        terminated_employees=("is_terminated", "sum"),
        avg_salary=("base_salary_aud", "mean"),
        avg_satisfaction=("job_satisfaction", "mean"),
        avg_sdg_score=("sdg_score", "mean")
    )
    .reset_index()
)

department_summary["attrition_rate_%"] = (
    department_summary["terminated_employees"] 
    / department_summary["employee_count"] 
    * 100
).round(2)

department_summary.to_csv(
    os.path.join(summary_folder, "department_summary.csv"), 
    index=False
)

# Workforce by office site
site_summary = (
    df_clean
    .groupby("office_site")
    .agg(
        employee_count=("employee_id", "count"),
        active_employees=("is_active", "sum"),
        terminated_employees=("is_terminated", "sum"),
        avg_satisfaction=("job_satisfaction", "mean"),
        avg_wellbeing=("wellbeing_index", "mean"),
        avg_sdg_score=("sdg_score", "mean")
    )
    .reset_index()
)

site_summary["attrition_rate_%"] = (
    site_summary["terminated_employees"] 
    / site_summary["employee_count"] 
    * 100
).round(2)

site_summary.to_csv(
    os.path.join(summary_folder, "site_summary.csv"), 
    index=False
)

# Pay equity by comparable role group and gender
pay_equity_summary = (
    df_clean
    .groupby(["comparable_role_group", "gender"])
    .agg(
        employee_count=("employee_id", "count"),
        avg_salary=("base_salary_aud", "mean"),
        median_salary=("base_salary_aud", "median"),
        avg_performance=("performance_rating", "mean"),
        avg_years_since_last_promotion=("years_since_last_promotion", "mean")
    )
    .reset_index()
)

pay_equity_summary.to_csv(
    os.path.join(summary_folder, "pay_equity_summary.csv"), 
    index=False
)

print("Summary tables exported successfully.")
```

```python
# ============================================================
# Create Career Background Long Table for Tableau Heatmap
# ============================================================
# If df_clean already exists from your cleaning notebook, use it directly.
# Otherwise load the Tableau-ready file:

career_fields = [
    "years_at_company",
    "years_in_current_role",
    "years_since_last_promotion",
    "years_with_current_manager"
]

career_long = df_clean.melt(
    id_vars=[
        "employee_id",
        "active_employee",
        "department",
        "job_role",
        "gender",
        "office_site",
        "age_band",
        "atsi_identification",
        "employee_status"
    ],
    value_vars=career_fields,
    var_name="career_metric",
    value_name="career_years"
)

career_long["career_metric"] = career_long["career_metric"].replace({
    "years_at_company": "Years at Company",
    "years_in_current_role": "Years in Current Role",
    "years_since_last_promotion": "Years Since Last Promotion",
    "years_with_current_manager": "Years With Current Manager"
})

def create_career_year_band(years):
    if years <= 1:
        return "0-1"
    elif years <= 3:
        return "2-3"
    elif years <= 5:
        return "4-5"
    elif years <= 10:
        return "6-10"
    else:
        return "10+"

career_long["career_year_band"] = career_long["career_years"].apply(create_career_year_band)

os.makedirs("data/processed", exist_ok=True)
career_long.to_csv("data/processed/hr_career_background_long.csv", index=False)

career_long.head()
```

```python
def clean_column_name(col):
    col = str(col).strip()
    col = col.lower()
    col = col.replace("(", "")
    col = col.replace(")", "")
    col = col.replace("/", "_")
    col = col.replace("-", "_")
    col = col.replace("&", "and")
    col = col.replace(" ", "_")
    col = col.replace("__", "_")
    return col

df.columns = [clean_column_name(col) for col in df.columns]

print("Cleaned columns:")
print(df.columns.tolist())

# ------------------------------------------------------------
# Step 2: Rename long/awkward columns if needed
# ------------------------------------------------------------

rename_dict = {
    "identifies_as_aboriginal_and_torres_strait_islander": "atsi_identification",
    "current_employee_status": "employee_status",
    "yearly_base_salary_aud": "base_salary_aud",
    "flexible_work_arrangement": "flexible_work"
}

df = df.rename(columns=rename_dict)

# ------------------------------------------------------------
# Step 3: Define score fields
# ------------------------------------------------------------

score_fields = [
    "inclusion_index",
    "wellbeing_index",
    "development_index",
    "work_life_balance",
    "job_satisfaction",
    "performance_rating",
    "sdg_score"
]

missing_fields = [col for col in score_fields if col not in df.columns]

if missing_fields:
    print("Missing score fields:", missing_fields)
    print("\nAvailable columns:")
    for col in df.columns:
        print(col)
    raise ValueError("Some score fields are still missing. Check column names above.")

# ------------------------------------------------------------
# Step 4: Create scorecard long table
# ------------------------------------------------------------

id_vars = [
    "employee_id",
    "active_employee",
    "terminated_flag",
    "department",
    "job_role",
    "job_level",
    "gender",
    "office_site",
    "office_type",
    "age_band",
    "atsi_identification",
    "employee_status"
]

# Keep only id_vars that exist to avoid errors
id_vars = [col for col in id_vars if col in df.columns]

scorecard_long = df.melt(
    id_vars=id_vars,
    value_vars=score_fields,
    var_name="score_metric",
    value_name="score_value"
)

# ------------------------------------------------------------
# Step 5: Clean metric labels for Tableau
# ------------------------------------------------------------

scorecard_long["score_metric"] = scorecard_long["score_metric"].replace({
    "inclusion_index": "Inclusion",
    "wellbeing_index": "Wellbeing",
    "development_index": "Development",
    "work_life_balance": "Work/Life Balance",
    "job_satisfaction": "Job Satisfaction",
    "performance_rating": "Performance",
    "sdg_score": "SDG Score"
})

# ------------------------------------------------------------
# Step 6: Create score band and sort order
# ------------------------------------------------------------

scorecard_long["score_value"] = scorecard_long["score_value"].round().astype(int)

scorecard_long["score_band"] = scorecard_long["score_value"].astype(str) + "/5"
scorecard_long["score_band_order"] = scorecard_long["score_value"]

metric_order = {
    "Inclusion": 1,
    "Wellbeing": 2,
    "Development": 3,
    "Work/Life Balance": 4,
    "Job Satisfaction": 5,
    "Performance": 6,
    "SDG Score": 7
}

scorecard_long["score_metric_order"] = scorecard_long["score_metric"].map(metric_order)

# ------------------------------------------------------------
# Step 7: Export
# ------------------------------------------------------------

os.makedirs("data/processed", exist_ok=True)

output_path = "data/processed/hr_people_scorecard_long.csv"
scorecard_long.to_csv(output_path, index=False)

print("Exported:", output_path)
print("Scorecard long shape:", scorecard_long.shape)

scorecard_long.head(10)
```
